```python
import os
```

```python
path = './contents'
contents = os.listdir(path)
```

```python
with open('test.html', 'r', encoding='utf-8') as file:
    html_content = file.read()
```

```python
url = "https://raw.githubusercontent.com/MINJU-B/KADaP_portal/main/"
for content in contents:
    print(content)
    img_url = url + content.replace('.png', '')
    print(img_url)
    html = html_content.replace("IMG_URL", img_url)

    with open(f'{content.replace('.png', '')}.html', 'w', encoding='utf-8') as f:
        f.write(html)
    break
```
